# File Operator

```python
print("----- WELCOME TO PERSONAL JOURNAL MANAGER -----\n")

class JournalManager():

    def file_create(self):
        
        try:
            f = open("journal.txt", "a")
            text = input("Enter Something:\n")
            f.write(text + "\n")
            print("Entry Added Successfully...")
            
        except Exception as e:
            print("Error while writing file:", e)
            
        finally:
            try:
                f.close()
            except:
                pass


    
    def file_all(self):

        try: 
            f = open("journal.txt", "r")
            data = f.readlines()

            if not data:
                print("File is Empty !")

            else:
                print("All Entries:")
                print("-" * 20)
                for line in data:
                    print(line.strip())
            
        except FileNotFoundError:
            print("File does not exist")
        
        except Exception as e:
            print("Error:", e)

        finally:
            try:
                f.close()
            except:
                pass

    def file_search(self):

        try:
            text = input("Enter Keyword to Search:-\n")
            f = open("journal.txt", "r")
            for i in f:
                if text in i:
                    print("Matching Entries:")
                    print("-" * 17)
                    print(i)
            else:
                print("No matching entry found !")
        
        except FileNotFoundError:
            print("File does not exist")

        except Exception as e:
            print("Error:", e)

        finally:
            try:
                f.close()
            except:
                pass

    def file_delete(self):

        try:
            confirm = input("Are you sure you want to delete all content? (yes/no): ")

            if confirm.lower() == "yes":
                f = open("journal.txt", "w")
                print("All content deleted successfully")
            else:
                print("Delete operation cancelled")

        except PermissionError:
            print("Permission denied")

        except Exception as e:
            print("Error:", e)

        finally:
            try:
                f.close()
            except:
                pass

F = JournalManager()

while True:
    
    try:
        choice = int(input("-------Menu-------\n1. Add a New Entry\n2. View All Entries\n3. Search for an Entry\n4. Delete All Entries\n5. Exit\n"))
    
        match choice:
    
            case 1:
                F.file_create()
            case 2:
                F.file_all()
            case 3:
                F.file_search()
            case 4:
                F.file_delete()
            case 5:
                print("Thank you ☺\nGood Bye !")
                break
            case _:
                print("Enter Valid Input")

    
    except ValueError:
        print("Please enter a valid number!")
```
