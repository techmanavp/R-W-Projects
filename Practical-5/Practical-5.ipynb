# OOP Wrapper

```python
print("----- Welcome to Employee Management System-----")


class Employee:
    
    def __init__(self, e_id, name, age, salary):
        self.__e_id = e_id
        self.name = name
        self.age = age
        self.__salary = salary
        

    def get_emp_id(self):
        return self.__e_id

    def get_emp_salary(self):
        return self.__salary

    def set_emp_id(self, e_id):
        self.__e_id = e_id

    def set_emp_salary(self, salary):
        self.__salary = salary

    def display(self):
        print("ID:", self.__e_id, "NAME:", self.name, "AGE:", self.age, "SALARY:", self.__salary, "₹")

    def __del__(self):
        print("DELETED...")
       


class Manager(Employee):
    
    def __init__(self, e_id, name, age, salary, department):
        super().__init__(e_id, name, age, salary)
        self.department = department
        

    def display(self):
        super().display()
        print("DEPARTMENT:", self.department)

class Developer(Employee):

    def __init__(self, e_id, name, age, salary, language):
        #Employee.__init__(self, e_id, name, age, salary)
        super().__init__(e_id, name, age, salary)
        self.language = language
        print("Developer has been added...")

    def display(self):
        super().display()
        print("LANGUAGE:", self.language)


l_emp = []
l_manager = []
l_developer = []

while True:
    choice = int(input("\n◄◄◄◄◄Choose an Option:◄◄◄◄◄\n1. Add an Employee\n2. Add a Manager\n3. Add a Developer\n4. Show Details\n5. Update Salary and ID\n6. Exit\n"))

    match choice:

        case 1: # CREATE AN EMPLOYEE

            E_id = input("Enter Employee ID:")
            E_name = input("Enter Employee Name:")
            E_age = int(input("Enter Employee Age:"))
            E_salary = int(input("Enter Employee Salary:"))

            E1 = Employee(E_id, E_name, E_age, E_salary)
            print("Employee has been added...")
            l_emp.append(E1)
            

        case 2: # CREATE A MANAGER

            M_id = input("Enter Manager ID:")
            M_name = input("Enter Manager Name:")
            M_age = int(input("Enter Manager Age:"))
            M_salary = int(input("Enter Manager Salary:"))
            M_deprt = input("Enter Department:")

            M1 = Manager(M_id, M_name, M_age, M_salary, M_deprt)
            print("Manager has been added...")
            l_manager.append(M1)

        case 3: #CREATE A DEVELOPER

            D_id = input("Enter Developer ID:")
            D_name = input("Enter Developer Name:")
            D_age = int(input("Enter Developer Age:"))
            D_salary = int(input("Enter Developer Salary:"))
            D_language = input("Enter Developer Language:")

            D1 = Developer(D_id, D_name, D_age, D_salary, D_language)
            print("Developerr has been added...")
            l_developer.append(D1)

        case 4:

            Q = int(input("\nWhom Details You Want To See\n1. Employee\n2. Manager\n3. Developer\n"))

            match Q:

                case 1:
                    for i in l_emp:
                        i.display()
                    print("-"*40)

                case 2:
                    for j in l_manager:
                        j.display()
                    print("-"*40)

                case 3:
                    for k in l_developer:
                        k.display()
                    print("-"*40)

        case 5:

            W = int(input("\nWhat Do You Want To Update:\n1. Update ID\n2. Update Salary\n"))

            match W:

                case 1:
                    
                    emp_id = input("\nEnter Employee ID:\n")

                    found = False
                    
                    for i in l_emp:
                        if i.get_emp_id() == emp_id:
                            new_id = input("Enter New ID:")
                            i.set_emp_id(new_id)
                            print("Updating...")
                            found = True
                            break

                    if not found:
                        print("Employee Not Found!")


                case 2:
                    
                    emp_id = input("\nEnter Employee ID:\n")

                    found = False
                    
                    for j in l_emp:
                        if j.get_emp_id() == emp_id:
                            new_salary = int(input("Enter New Salary:"))
                            j.set_emp_salary(new_salary)
                            print("Updating...")
                            found = True
                            break

                    if not found:
                        print("Employee Not Found!")

                case _:
                    print("Enter Valid Input !")
        
        
        case 6:

            print("Exiting:)")
            break

        case _:

            print("Enter Valid Input")

                                
```
