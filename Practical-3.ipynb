# Collection Manipulator

```python
print("Welcome to Student data Organizer !")

students = []
subject_set = set()

while True:
    print()
    print("Select an Option:\n""1. Add Student\n""2. Display Students""\n3. Update Student Info""\n4. Delete Student""\n5. Subject List""\n6. Exit")
    print()
    ch = int(input("Enter Your Choice:"))
    print()
    
##################################################################################################################################################    
    match ch:
        case 1:
            n = int(input("How Many Student Do You Want to Add:"))
            print("Please Fill The Details")
            print()
            
            for i in range(n):
                name = input("Enter Your Name:")
                en_no = int(input("Enter Your Enrollment Number:"))
                dob = input("Enter Your DOB (DD-MM-YYYY):")
                grade = input("Enter Your Grade:")
                sub = input("Enter Your Subject:").split(",")


                info = (en_no, dob)
                
                
                # student = {"Name":name,
                #       "En_no":en_no,
                #       "DOB":dob,
                #       "Grade":grade,
                #       "Subjects":sub}
                student = {"Info":info,
                           "Name":name,
                           "Grade":grade,
                            "Subjects":sub}
                
                
                subject_set.update(sub)

                
                students.append(student)
                print("\nStudent Add Successfully..")

#################################################################################################################################################
              

        case 2:
            print("All Students Records")
            for i in students:
                print(i)

#################################################################################################################################################
        case 3:
            num = int(input("Enter Enrollment Number to Search Student:"))

            for i in students:
                if i["Info"][0] == num:
                    print("Student Found")
                    print("What Do You Want to Update""\n1. Name, 2. Grade, 3. Subject")
                    print()

                    ch3 = int(input("Enter Your Choice:"))

                    if ch3 == 1:
                        new_name =  input("Enter New Name:")
                        i["Name"] = new_name
                        print("Name Update Successfully..")

                    elif ch3 == 2:
                        new_grade = input("Enter New Grade:")
                        i["Grade"] = new_grade
                        print("Grade Update Successfully..")

                    elif ch3 == 3:
                        new_sub = input("Enter New Subjects:").split(",")
                        i["Subjects"] = new_sub
                        subject_set.update(new_sub)
                        print("Subjects Updated Successfully..")

                    break
                
                
                else:
                    print("Enter Valid Enrollmet Number")
        
        
###############################################################################################################################################    
        case 4:
            dl_student = int(input("Enter Enrollment Number to Delete Student Entry:"))

            for i in students:
                if i["Info"][0] == dl_student:
                    students.remove(i)
                    print("Student Entry Deleted Successfully..")
                    break

            else:
                    print("Enter Valid Enrollmet Number")


        case 5:
    
            print("Subjects Offered:")

            for s in subject_set:
                print(s)
###############################################################################################################################################

        case 6:
            print("Exit")
            break
```
